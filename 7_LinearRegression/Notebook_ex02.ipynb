```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

X, y, coef = make_regression(n_samples=100,
                         n_features=1,
                         n_informative=1,
                         noise=10,
                         coef=True,
                         random_state=0,
                         bias=100.0)
```

```python
plt.scatter(X, y)
plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression Data')
print("Question 1")
plt.show()
```

```python
# Fit the model
model = LinearRegression()
model.fit(X, y)

# Get the coefficients and the intercept
coef = model.coef_
intercept = model.intercept_

print("Question 2")
print(f'y = {coef[0]} * x + {intercept}')
```

```python
# Fit a line to the data
coef, intercept = np.polyfit(X.flatten(), y, 1)

# Create a sequence of x values
x_line = np.linspace(X.min(), X.max(), 100)

# Create a sequence of y values for the fitted line
y_line = coef * x_line + intercept

# Plot the data
plt.scatter(X, y)

# Plot the fitted line
plt.plot(x_line, y_line, color='red')

plt.xlabel('Feature')
plt.ylabel('Target')
plt.title('Linear Regression Data with Fitted Line')

print("Question 3")
plt.show()
```

```python
# Predict the target values
pred = model.predict(X.reshape(-1, 1))

print("Question 4")
print(pred[:10])
```

```python
def compute_mse(y_true, y_pred):
    mse = ((y_true - y_pred)**2).mean()
    return mse

X, y, coef = make_regression(n_samples=100,
                         n_features=1,
                         n_informative=1,
                         noise=10,
                         coef=True,
                         random_state=0,
                         bias=100.0)
model.fit(X,y)

# Predict on X
pred = model.predict(X)

# Compute the MSE
mse = compute_mse(y,pred)
#The difference is likely due to floating-point precision in Python. 
#It's common to have very slight variations in floating-point calculations. 
#The difference between 114.17148616819486 and 114.17148616819485 is extremely small and should not significantly impact your analysis
print("Question 5")
print(f'Mean Squared Error (MSE): {mse}, built in functin from sklern gives {mean_squared_error(y, pred)}')
```
