```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def sigmoid1(x):
    return 1.0 / (1.0 + np.exp(-(0.5*x + 3)))

def sigmoid2(x):
    return 1.0 / (1.0 + np.exp(-(5*x + 11)))
```

```python
x = np.linspace(-10, 10, 1000)
y_sigmoid = sigmoid(x)
y_sigmoid1 = sigmoid1(x)
y_sigmoid2 = sigmoid2(x)
```

```python
plt.plot(x, y_sigmoid, label='sigmoid(x)')
plt.plot(x, y_sigmoid1, label='sigmoid1(x)')
plt.plot(x, y_sigmoid2, label='sigmoid2(x)')
plt.axhline(y=0.5, color='r', linestyle='--')  # Add a line representing the probability=0.5
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.title('Sigmoid Functions')
#plt.legend()
plt.show()
```
